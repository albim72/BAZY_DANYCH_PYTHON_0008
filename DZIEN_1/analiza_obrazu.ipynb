# Operacje na obrazach - negatyw i progowanie

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#tworzymy sztuczny obraz 8x8
np.random.seed(0)
image = np.random.randint(0,256,(8,8),dtype=np.uint8)
```

```python
#negatyw
negative = 255-image
```

```python
#progowanie czarno-białe przy progu 128
threshold = (image>128)*255 #True->255, False->0
```

```python
#wizualizacja
fig,axs = plt.subplots(1,3,figsize=(6,2))
axs[0].imshow(image,cmap='gray');axs[0].set_title("Oryginał")
axs[1].imshow(negative,cmap='gray');axs[1].set_title("Negatyw")
axs[2].imshow(threshold,cmap='gray');axs[2].set_title("Progowanie")
for ax in axs:ax.axis("off")
plt.tight_layout()
plt.show()
```
